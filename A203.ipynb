```python
import pandas as pd

# Carregar a base (ajuste o caminho se necessário)
df = pd.read_csv("train (1).csv")

# INSPEÇÃO INICIAL
print("PRIMEIRAS LINHAS")
display(df.head())  # Visualização inicial da base

print("DIMENSÃO DA BASE")
print(df.shape)  # (n_linhas, n_colunas)

print("INFORMAÇÕES GERAIS")
df.info()  # Tipos de dados e quantidade de valores não nulos

print("ESTATÍSTICAS DESCRITIVAS")
display(df.describe())  # Estatísticas das variáveis numéricas

# RESPOSTAS ÀS PERGUNTAS
# 1) Quantas observações e variáveis?
n_linhas, n_colunas = df.shape
print(f"\nA base possui {n_linhas} observações e {n_colunas} variáveis.")

# 2) Qual é a variável alvo?
print("\nVariável alvo esperada: SalePrice")
if "SalePrice" in df.columns:
    print("A variável alvo 'SalePrice' está presente na base.")
else:
    print("Verifique o nome da variável alvo.")

# 3) Existem variáveis categóricas?
variaveis_categoricas = df.select_dtypes(include="object").columns
print(f"\nQuantidade de variáveis categóricas: {len(variaveis_categoricas)}")

# 4) Existem colunas com muitos valores ausentes?
missing = df.isna().sum()
missing = missing[missing > 0]

print(f"\nQuantidade de colunas com valores ausentes: {len(missing)}")

print("\nColunas com mais de 20% de valores ausentes:")
display((df.isna().mean() * 100)[(df.isna().mean() * 100) > 20])
```

Parte 2 – Diagnóstico de valores ausentes

Calcule a quantidade e o percentual de valores ausentes por coluna.

Identifique:
as 5 colunas com maior percentual de valores ausentes;
se a variável alvo possui valores ausentes.

Perguntas reflexivas:
É justificável remover colunas com alto percentual de ausência?
A ausência pode carregar informação (ex.: ausência de piscina)?

Entrega esperada:
Tabela com percentuais de ausência e justificativa técnica para remoção ou imputação.

```python
# Cálculo da quantidade de valores ausentes
missing_abs = df.isna().sum()

# Cálculo do percentual
missing_pct = df.isna().mean() * 100

# Construção da tabela organizada
missing_table = pd.DataFrame({
    "Valores Ausentes": missing_abs,
    "Percentual (%)": missing_pct
}).sort_values("Percentual (%)", ascending=False)

# Exibir as 5 colunas com maior percentual
top5_missing = missing_table.head(5)

display(top5_missing)

# Verificar se o target possui valores ausentes
print("Valores ausentes na variável alvo:")
print(df["SalePrice"].isna().sum())
```

Parte 3 – Variáveis categóricas e numéricas

Separe variáveis numéricas e categóricas.

Calcule:
número de categorias distintas por variável categórica;
correlação das variáveis numéricas com o target.

Perguntas orientadoras:
Existem variáveis com alta cardinalidade?
Alguma variável apresenta correlação muito elevada com o preço?
Existe risco de multicolinearidade?

```python
# Separação por tipo de dado
variaveis_numericas = df.select_dtypes(include=["int64", "float64"]).columns
variaveis_categoricas = df.select_dtypes(include=["object"]).columns

print("Quantidade de variáveis numéricas:", len(variaveis_numericas))
print("Quantidade de variáveis categóricas:", len(variaveis_categoricas))
```

```python
# Número de categorias distintas
cardinalidade = df[variaveis_categoricas].nunique().sort_values(ascending=False)

display(cardinalidade.head(10))  # Mostrar as 10 com maior cardinalidade
```

```python
# Correlação com o target
correlacoes = df[variaveis_numericas].corr()["SalePrice"].sort_values(ascending=False)

display(correlacoes.head(10))   # Maiores correlações positivas
display(correlacoes.tail(10))   # Maiores correlações negativas
```

```python
# Matriz de correlação completa
matriz_corr = df[variaveis_numericas].corr()

# Identificar pares com correlação alta
corr_alta = (
    matriz_corr.abs()
    .where(lambda x: x > 0.8)
    .stack()
    .reset_index()
)

corr_alta = corr_alta[corr_alta["level_0"] != corr_alta["level_1"]]

display(corr_alta.head(10))
```

Vamos fazer a Regressão Linear?

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Carregar base
df = pd.read_csv("train (1)  Resumida.csv")

# Definir variável explicativa e alvo
X = df[["OverallQual"]]
y = df["SalePrice"]

# Criar e ajustar modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Mostrar equação
print("Intercepto:", modelo.intercept_)
print("Coeficiente:", modelo.coef_[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, modelo.predict(X), color="red")
plt.xlabel("OverallQual")
plt.ylabel("SalePrice")
plt.show()
```
