```python
import pandas as pd


dados = {
    "Categoria": ["Iluminacao", "Ar-Condicionado", "Outros", "Total"],
    "Consumo_kWh": [120, 200, 530, 850],
    "Desperdicio_kWh": [120, 200, 0, 320]
}

df = pd.DataFrame(dados)
df.to_csv("consumo_energia.csv", index=False)

print("Arquivo 'consumo_energia.csv' criado com sucesso!")
```

```python
df = pd.read_csv("consumo_energia.csv")

print("Resumo dos dados de consumo:")
print(df.describe())

total_consumo = df['Consumo_kWh'].sum()
total_desperdicio = df['Desperdicio_kWh'].sum()

print(f"\nConsumo total: {total_consumo} kWh/mês")
print(f"Desperdício total: {total_desperdicio} kWh/mês")
```

```python
economia_iluminacao = df.loc[df['Categoria']=='Iluminacao', 'Desperdicio_kWh'].values[0] * 0.7
economia_ar = df.loc[df['Categoria']=='Ar-Condicionado', 'Desperdicio_kWh'].values[0] * 0.5

economia_total = economia_iluminacao + economia_ar

print(f"Economia estimada:")
print(f" - Iluminação: {economia_iluminacao} kWh/mês")
print(f" - Ar-condicionado: {economia_ar} kWh/mês")
print(f" - Total: {economia_total} kWh/mês")
```
